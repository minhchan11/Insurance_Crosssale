```python
def convert_height(string_height):
    splitted = string_height.split(',')
    if splitted[2]=='FI':
        if splitted[1] == '':
            splitted[1] = '0'
        return float(ast.literal_eval(splitted[0]))*30.48 + float(ast.literal_eval(splitted[1]))*2.54
    else:
        return ast.literal_eval(splitted[0])
```

```python
def convert_weight(string_weight):
    splitted = string_weight.split(' ')
    if len(splitted)==2:
        if splitted[1]=='P':
            return float(ast.literal_eval(splitted[0]))*0.453592
        elif splitted[1]=='K':
            return float(ast.literal_eval(splitted[0]))
        else: 
            return 0
    else:
        return float(ast.literal_eval(splitted[0]))
```

# CLEAN DATASET

## * There is a need to standardize all string and convert measurements into the same unit

```python
df.fillna('0',inplace=True)
```

```python
df.columns
```

```python
for i in ['PROPOSAL_SUBMISSION_DATE','ISSUE_DATE','ST_DATE']:
    df[i] = df[i].apply(lambda x: datetime.strptime(x, '%m/%d/%Y') if x !='0' else datetime.min)
```

```python
df['PROPOSAL_TO_ISSUE'] = df['PROPOSAL_SUBMISSION_DATE'] - df['ISSUE_DATE']
```

```python
df['ISSUE_TO_ST'] = df['ISSUE_DATE'] - df['ST_DATE']
```

```python
for i in ['ISSUE_TO_ST','PROPOSAL_TO_ISSUE']:
    df[i] = df[i].apply(lambda x: dt.timedelta(0) if x < dt.timedelta(0) or x>dt.timedelta(1000) else x)
```

```python
df['HEIGHT_CM'] = df['HEIGHT'].apply(lambda x: int(convert_height(x)) if x !='0' else x)
```

```python
df['WEIGHT_KG'] = df['WEIGHT'].apply(lambda x: int(convert_weight(x)) if x !='0' else x)
```

```python
df['POL_ST'] = df['POL_ST'].apply(lambda x: x.lower())
```

```python
def find_person_id(id1,id2):
    if id2 == '0':
        return id1
    else:
        return id2
```

```python
df['PERSON_ID'] = df.apply(lambda x: find_person_id(x['INSURED_ID'],x['BUYER_ID']),axis=1)
```

```python
for i in ['INSURED_AMT','COVER_CODE','PRODUCT_CODE','PREMIUM']:
    df[i] = df[i].apply(lambda x: float(x.replace(',','')))
```
